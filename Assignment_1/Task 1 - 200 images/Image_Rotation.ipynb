```python
# Step 3: Import necessary libraries
import os
import random
from PIL import Image

# Step 4: Define the path to your images folder in Google Drive
image_folder_path = '/content/drive/MyDrive/textual_news_image_dataset'  # Change this to your actual folder path
```

```python
# Step 5: Function to rotate image and save it with a new name
def rotate_image(image_path, count):
    # Open an image file
    with Image.open(image_path) as img:
        # Generate a random angle between 0 and 360 degrees
        angle = random.randint(0, 360)

        # Rotate the image
        rotated_img = img.rotate(angle, expand=True)

        # Get the original file extension
        file_extension = os.path.splitext(image_path)[1]

        # Get the base name of the file without extension
        base_name = os.path.splitext(os.path.basename(image_path))[0]

        # Create a new name for the rotated image
        new_image_path = os.path.join(os.path.dirname(image_path), f"{base_name}_rotated_{count}{file_extension}")

        # Save the rotated image with the new name
        rotated_img.save(new_image_path)
        print(f"Rotated and saved: {new_image_path} (Angle: {angle} degrees)")

# Step 6: Process each image in the folder
count = 1  # Initialize a counter
for image_file in os.listdir(image_folder_path):
    if image_file.endswith(('.png', '.jpg', '.jpeg','PNG')):  # Add other formats if needed
        image_path = os.path.join(image_folder_path, image_file)
        print(f"Processing: {image_file}")
        rotate_image(image_path, count)
        count += 1  # Increment the counter for each image
```
