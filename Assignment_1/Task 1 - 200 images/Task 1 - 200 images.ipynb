```python
import os
import pytesseract
from PIL import Image

# For Windows, specify the path to your Tesseract OCR executable:
# Uncomment and modify the following line if you're on Windows:
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Folder path containing your images
image_folder = '/content/drive/MyDrive/tech_image/news'

# Path to save the text file
output_text_file = 'extracted_text.txt'

# Open the text file in write mode
with open(output_text_file, 'w') as file:
    # Loop through the image folder
    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            img_path = os.path.join(image_folder, filename)
            try:
                # Debugging: Check if the file path is correct
                print(f"Processing {img_path}")

                # Open the image file
                img = Image.open(img_path)

                # Debugging: Check if the image is opened properly
                print(f"Opened image {filename}")

                # Use Tesseract to extract text from the image
                text = pytesseract.image_to_string(img)

                # Debugging: Check the extracted text
                print(f"Extracted text from {filename}: {text}")

                # Write the filename and the extracted text into the text file
                # file.write(f"Text from {filename}:\n")
                file.write(text + "\n\n")

            except Exception as e:
                print(f"Could not process image {filename}: {e}")

print(f"Text extraction complete! Extracted text is saved in {output_text_file}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Path to your extracted text file
text_file_path = '/content/extracted_text.txt'

# Read the content of the text file
with open(text_file_path, 'r') as file:
    text = file.read()

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No axes needed for the word cloud
plt.show
```
